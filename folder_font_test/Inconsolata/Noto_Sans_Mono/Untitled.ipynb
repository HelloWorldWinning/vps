```python
import os
from fontTools.ttLib import TTFont

ttf_path = 'static'
files = os.listdir(ttf_path)

for file_name in files:
    if file_name.endswith('.ttf'):
        file_path = os.path.join(ttf_path, file_name)
        font = TTFont(file_path)

        print(f"Information for {file_name}:")
        name_table = font['name']
        
        # Iterate through the name records in the name table
        for record in name_table.names:
            # Decode the string based on its encoding
            if record.isUnicode():
                text = record.string.decode('utf-16-be')
            else:
                text = record.string.decode('latin-1')
            
            print(f"{record.nameID}: {text}")

        # Optionally, list other table names or details here
        print("\nTable names in the font:")
        print(list(font.keys()))
        print("---------------------------------------------------------------------------")
```
