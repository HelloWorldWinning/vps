```python
import os
from fontTools.ttLib import TTFont

# Specify the directory containing the TTF files
# ttf_path = '.'


ttf_path = 'static'
files = os.listdir(ttf_path)

# Iterate through all files in the specified directory
for file_name in files:
    # Check if the file is a TTF file
    if file_name.endswith('.ttf'):
        # Construct the full path to the file
        file_path = os.path.join(ttf_path, file_name)
        
        # Load the TTF file
        font = TTFont(file_path)
        print(f"Variations in {file_name}:")

        # Access the 'name' table
        name_table = font['name']
        for record in name_table.names:
            # Check if the record is the subfamily name (usually contains style information)
            if record.nameID == 2:
                # Decode and print the subfamily name (style)
                style_name = record.toUnicode()
                print("="*20)
                print(f"  Style: {style_name}")
```

```python
import os
from fontTools.ttLib import TTFont

# Path to the directory containing TTF files
ttf_path = 'static'
files = os.listdir(ttf_path)
first_file = files[0] if files else None

if first_file and first_file.endswith('.ttf'):
    # Construct the full path to the file
    file_path = os.path.join(ttf_path, first_file)
    
    # Load the font file
    font = TTFont(file_path)
    
    # Accessing font information
    print(f"Information for {first_file}:")
    
    # Access the 'name' table to get font information
    name_table = font['name']
    for record in name_table.names:
        text = record.string.decode('utf-16-be') if record.isUnicode() else record.string.decode('latin-1')
        print(f"{record.nameID} ({record.getName(record.nameID, record.platformID, record.platEncID, record.langID)}): {text}")
    
    # You can also access other tables and information as needed
    # For example, listing the table names in the font
    print("\nTable names in the font:")
    print(font.keys())
```

```python
import os
from fontTools.ttLib import TTFont

ttf_path = 'static'
files = os.listdir(ttf_path)

for file_name in files:
    if file_name.endswith('.ttf'):
        file_path = os.path.join(ttf_path, file_name)
        font = TTFont(file_path)

        print(f"Information for {file_name}:")
        name_table = font['name']
        
        # Iterate through the name records in the name table
        for record in name_table.names:
            # Decode the string based on its encoding
            if record.isUnicode():
                text = record.string.decode('utf-16-be')
            else:
                text = record.string.decode('latin-1')
            
            print(f"{record.nameID}: {text}")

        # Optionally, list other table names or details here
        print("\nTable names in the font:")
        print(list(font.keys()))
        print("---------------------------------------------------------------------------")
```
